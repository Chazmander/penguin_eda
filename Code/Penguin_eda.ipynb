```python
# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'penguins_size.csv'
penguins = pd.read_csv(file_path)
penguins.head()
```

```python
# Check for missing values
penguins.isnull().sum()
```

```python
# Summary statistics of the dataset
penguins.describe()
```

```python
# Visualize the distribution of species
sns.countplot(data=penguins, x='species')
plt.title('Distribution of Penguin Species')
plt.show()
```

```python
# Pairplot to visualize relationships between numerical features
sns.pairplot(penguins, hue='species')
plt.show()
```

```python
# Check the unique values in the 'species' column
penguins['species'].value_counts()
```

```python
# Visualize the distribution of body mass
sns.histplot(data=penguins, x='body_mass_g', kde=True, hue='species')
plt.title('Distribution of Body Mass by Species')
plt.show()
```

```python
# Boxplot of culmen length by species
sns.boxplot(data=penguins, x='species', y='culmen_length_mm')
plt.title('Culmen Length by Species')
plt.show()
```

```python
# Correlation heatmap of numerical features
correlation_matrix = penguins.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
